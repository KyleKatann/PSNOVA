```python
from pandas import*
from glob import*

def printtag(line,tmp):
    tag="        <tr>"
    if not line[-1]:
        line[-1]="-"

    for item in zip(*tmp):
        tag+='<th>'+str(item[0])
        for name in item[1:]:
            if name:
                tag+='<br>'+name
        tag+='</th>'
    tag+="</tr>"

    if tmp:
        print(tag)


for name in sorted(glob("./excel/*.xlsx")):
    df=read_excel(name,header=None).fillna("").as_matrix()
    
    print("<details>")
    print("    <summary>"+name.split("_")[1][:-5]+"</summary>")
    print('    <table border="1" style="border-collapse: collapse">')
    
    tag="        <tr>"
    for h in df[0]:
        tag+='<th bgcolor="#87cefa">'+h+'</th>'
    tag+="</tr>"
    print(tag)
    
    tmp=[]
    for line in df[1:]:
        if line[0]:
            printtag(line,tmp)
            
            tmp=[]
            tmp.append(list(line))
        else:
            tmp.append(list(line))
            
    printtag(line,tmp)
    
    print("    </table>")
    print("</details>")
    
    
    
    
```
